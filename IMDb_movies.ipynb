# **6. Data Pre-Processing**
---
- Here we will perform **Data Preprocessing** on our dataset, to make the data usable for **EDA**.


  - Handling missing data,

  - Handling redundant data,

  - Handling inconsistent data,

  - Handling outliers,

  - Handling typos

```python
movies_df.to_excel('Movies_data.xlsx')                                         # Exporting the original Data
movies_df.head()
```

```python
movies_df.columns = map(str.lower, movies_df.columns)                 # map() function returns a list of results after applying the given function
                                                                      # to each item of a given iterable.
movies_df.dtypes                                                      # Confirm that all variable names are now lower case
```

```python
#missing data
total = movies_df.isnull().sum().sort_values(ascending=False)
total
```

```python
#replacing missing values of of metascore column by mean
movies_df[movies_df["metascore"].isnull()]
movies_df['metascore'] = movies_df['metascore'].fillna(movies_df['metascore'].mean())
```

```python
#Creating a new column as rating group, since raing has better correlation with revenue than metscore
movies_df['rating_group'] = pd.cut(movies_df["rating"].dropna(), [0,1.9,3.9,5.9,7.9,10],
                                labels=['0-1.9','2-3.9','4-5.9','6-7.9','8+'], 
                                include_lowest=True)
```

```python
#replacing missing values of of revenue column by mean after grouping the data rating wise
movies_df['revenue (millions)'].fillna(movies_df.groupby(['rating_group'])['revenue (millions)'].transform('median'),inplace=True)

movies_df.groupby('rating_group')['revenue (millions)'].agg(['median','mean'])
```

```python
#changing the year column to year formate
movies_df['year'] = pd.to_datetime(movies_df['year'], format='%Y').dt.year
movies_df['year'].dtype
```

```python
# Visualizing the Outliers using seaborn.boxplot for revenue coulmn
    fig_dims=(15,3.5)
    fig, ax1 = plt.subplots(figsize=fig_dims)
    #sns.set_style("whitegrid")
    ax2 = sns.boxplot(x='revenue (millions)', data=movies_df)
    ax2.set_xlabel("revenue (millions)")
    ax2. xaxis. label. set_size(15)
#ax.set(xlim=(100,600000))
```

```python
#Finding the list of unique genres
genres = movies_df['genre'].str.split(',').explode().tolist()
genres = list(set(genres))
genres
```

- Since the **genre** column contain a list of genres on which we can not do much analysis,Hence we will split them into three cloumns by considering the first genre to be a **primary genre** of the movie.

```python
#Creating new dataframe by spliting genre column
genre_df = movies_df['genre'].str.split(',', expand=True)

genre_df.rename(columns = {0: 'genre_primary', 1: 'genre2', 2:'genre3'}, inplace = True)
genre_df
```

```python
#using concat to merge genre daframe with the main dataframe
movies_df_new = pd.concat([movies_df, genre_df], axis = 1)
movies_df_new
```

- We might has missed some point by splitting the **genre column**, hence we will create a new data from by grouping all 3 genre columns to count each genre in list as a main genre.

```python
#Creating 3 dataframe by grouping each genre column
df1 = movies_df_new.groupby('genre_primary').agg(['sum','mean'])
df2 = movies_df_new.groupby('genre2').agg(['sum','mean'])
df3 = movies_df_new.groupby('genre3').agg(['sum','mean'])
```

```python
#Adding all three datafarme to created a single dataframe
df_extra = df1.add(df2, fill_value = 0)
df_unique_genre = df_extra.add(df3, fill_value = 0)
```

### **9.1 Conclusion**

1. The market for the film industry is increasing rapidly, as we can see there is a 500% growth in terms of movies released per year in the year 2016 as compared to 2006 which means people like to watch movies for their entertainment. 
2. Across all genre critic ratings of a movie quite close to those of the audience ratings.
3. Most movies last between 90 to 150 minutes.
4. The public tends to vote more and rate more when they like a movie, people avoid rating when they do not like a movie. The behavior can be seen in 2012 when movies made high revenue hence both ratings and avg votes were higher whereas votes are the lowest in 2016, which confirms our understanding. 
5. People more likely to rate a movie higher which has a length between 120 â€“ 170 minutes, critics (Metascore) also have quite a similar trend.  
6. Adventure is the most popular genre with the highest gross revenue as well as the highest average revenue over the last 10 years followed by Action and Comedy. Since genre was given in a group of a minimum of one to three it may give us a false impression as for a movie there may be multiple genres counted.  
- For movies with multiple genres if we assume the first genre as the primary genre. 
7. Animation movies have been most successful with the highest average revenue overall followed by Action and Adventure. 
8. Drama and comedy have a significant release in the past 10 years but their revenue generation is quite low.  
9. Animation, Biography, Drama, and Adventure are top-rated movies by both public and critics where Si-Fi and Horror are the lowest rated. 
10. The films that rated well by the public and critics has earned really well at the box office.

### **9.2 Actionable Insights**

To improve the quality and add new services, the company shall consider the following actionable insights:

1. The company can release more movies having runtime between 130 - 180 minutes to increase the revenue.  
2. They can release more Animation and Adventure genres, as the overall count of these genres is quite low being the most successful genre.
